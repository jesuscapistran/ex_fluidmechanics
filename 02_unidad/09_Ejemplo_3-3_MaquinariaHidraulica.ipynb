**Ejemplo 3-2**

Considere un gato hidráulico que se utiliza en un taller de reparación de automóviles, como se muestra en la figura 3-12. Los pistones tienen un área de $A1 = 0.8 cm^2$ y $A2 = 0.04 m^2$. Se bombea aceite hidráulico con una gravedad específica de 0.870, a medida que el pequeño pistón en el lado izquierdo se empuja hacia arriba y hacia abajo, con lo cual se eleva lentamente el pistón más grande del lado derecho.

- ¿Cuál es la fuerza (F1) necesaria para levantar una camioneta de 3 toneladas a una altura de 2 m? 

$$P_1 = P_2 + \rho g h $$

$$ \frac{F_1}{A_1} = \frac{F_2}{A_2} + \rho g h  $$

Ecuación que se debe programar ¿Cuál es el valor de F1? 
$$  F_1 = \frac{F_2 A_1}{A_2} + \rho g h A_1 $$




```python
# Librearias necesarias 
```

```python
# Declarar constantes y variables 

A1 = 0.8        # [cm2]
A1 = A1/(100**2)# [m2] Conversión a m2 - https://www.google.com/search?q=0.8+cm2+a+m2
A1 = 8e-5       # [m2] 
A2 = 0.04
GE = 0.870      # Gravedad Especifica
rho= 0.870*1000 # [Kg/m3] Rho = (GE)*(densidad agua)
g  = 9.81       # [m/s2]
h  = 2          # [m]  Altura de levantamiento 
w  = 3000       # [kg] Masa de camioneta , 3 Toneladas
F2 = w*g        # [(kg m)/s2] = [N] La fuerza F2 es igual al peso de la camioneta 
 
```

```python
# Ecuación a resolver 

F1 = (F2*A1)/A2 + rho*g*h*A1
F1 = round(F1,2)
F1
```

```python
print(f'La F1 necesaria para levantar una camioneta de {w} kg es {F1} N')
```
