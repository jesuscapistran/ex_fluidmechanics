### **Ejercicio 4-1** Un campo de velocidad bidimensional estacionario 

Se da un campo de velocidad estacionario, incompresible y bidimensional por: 

$$ \vec{V} = (u,v) = (0.5+0.8x)\vec{i} + (1.5-0.8y)\vec{j}$$

donde las coordenadas $x$ y $y$ se dan en metros y la magnitud de la velocidad está en $m/s$. 


 - Determínese si existen puntos de estancamiento en este campo de flujo y, si es así, ¿en dónde?

 - Trace un esquema de vectores de velocidades en varias ubicaciones en el dominio, entre $x = -2\:m$ hasta $2 \:m$  y $ y = 0$ hasta $5 \: m$; describa cualitativamente el campo de flujo. 

Nota: Un **punto de estancamiento** se define como un punto en el campo de flujo en donde la velocidad es idénticamente cero. 


```python
# Importar librerias 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Tutorial: https://youtu.be/u0hfen9VkJs 
# Definir el mallado del campo de vectores 

x_range = np.linspace(-3,3, 10) 
y_range = np.linspace( -1,5, 10)

x, y = np.meshgrid(x_range, y_range)


```

```python
# Introducir el campo de velocidad (Estacionario, incompresible, bidimensional)

Vi = 0.5 + 0.8*x
Vj = 1.5 - 0.8*y

```

```python
#Graficar el campo de velocidad (vectorial)

plt.quiver(x, y, Vi, Vj, pivot ='middle')
plt.axis('scaled')
plt.show()
```

```python
# Mejorar la grafica de campo de vectores de velocidad 
plt.figure(figsize=(5,5))
V = plt.quiver(x, y, Vi, Vj, pivot ='middle') # Para graficar campo de vectores
plt.quiverkey(V, 0.85, 1.03, 10, '10 m/s', labelpos ='E')

plt.title('Escala:')
plt.xlabel('x', fontsize = 15) #Nombre del eje x 
plt.ylabel('y', fontsize = 15) #Nombre del eje y
plt.axis('scaled') 
plt.show()
```

### Ejemplo 4-4: Lineas de Corriente en el plano XY: Una solución análitica

Para el campo bidimensional estacionario e incompresible de velocidad del ejemplo 4-1, trace la gráfica de varias **líneas de corriente** en la mitad derecha del flujo **(x > 0)** y haga una comparación con los vectores velocidad trazados en la figura 4-4.

```python
# Importar librerias 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir el mallado del campo de vectores 

x1_range = np.linspace(0,5.6, 8) 
y1_range = np.linspace( -1,5.6, 8)

x1, y1 = np.meshgrid(x1_range, y1_range)

# Introducir el campo de velocidad (Estacionario, incompresible, bidimensional)

Vi_1 = 0.5 + 0.8*x1
Vj_1 = 1.5 - 0.8*y1


# Graficar en la mitad derecha del flujo: x > 0

plt.figure(figsize=(5,5))
V = plt.quiver(x1, y1, Vi_1, Vj_1, pivot ='tail') # Para graficar campo de vectores
plt.quiverkey(V, 0.85, 1.03, 10, '10 m/s', labelpos ='E')

plt.title('Escala:')
plt.xlabel('x', fontsize = 15) #Nombre del eje x 
plt.ylabel('y', fontsize = 15) #Nombre del eje y
plt.axis('scaled') 
plt.grid()
plt.show()
```

La linea de corriente es tangente al vector velocidad , por lo tanto
se debe calcular: $dy/dx = v/u$. En otras palabras debemos resolver la siguiente ecuación diferencial (primer orden): 


$$ \frac{dy}{dx}  = \frac{v}{u} =\frac{1.5-0.8y}{0.5+0.8x}$$

- Tutorial ODE: https://www.youtube.com/watch?v=LbcjNGE5gkM 
- SciPy integrate: https://docs.scipy.org/doc/scipy/tutorial/integrate.html

```python
# Importar libreria Symbolic Python 
from scipy import integrate
```

```python
#Definir la función (Ecuación diferencial) a evaluar 

def funcion(y2,x2):
    return (1.5-0.8*y2)/(0.5 + 0.8*x2)

# Definir condiciones iniciales  

y0 = np.linspace(-20,20,11)
#y0 = 3

# Definir el intervalo de evaluación 

x2 = np.linspace(0,6,100) 

# Solución (linea de corriente) de la ecuación diferencial cuando yo = 3

linea_0 = integrate.odeint(funcion,y0,x2)
```

```python
print(y0)
```

```python
# Graficar campo de velocidad

plt.figure(figsize=(5,5))
V = plt.quiver(x1, y1, Vi_1, Vj_1, pivot ='tail') # Para graficar campo de vectores
plt.quiverkey(V, 0.85, 1.03, 10, '10 m/s', labelpos ='E')

# Grafica de las lineas de corriente 
plt.plot(x2,linea_0)

# Elementos de la grafica: Titulos
plt.title('Escala:')
plt.xlabel('x', fontsize = 15) #Nombre del eje x 
plt.ylabel('y', fontsize = 15) #Nombre del eje y
plt.axis('scaled') 

# Limite de los ejes 
plt.xlim([-0, 5.6])
plt.ylim([-1, 5.6])

plt.grid()
plt.show()
```
